```python
import numpy as np
import sombrero
```

```python
t = np.arange(0, 2*np.pi*50+1, 0.001)
r0 = np.array([0, 0])
r = sombrero.runge_kutta_4thOrd(r0, t, F=0.4)
x = np.array([i[0] for i in r])
y = np.array([i[1] for i in r])
sombrero.plot(t, x, label="x(t)", xlabel="t", ylabel="x", title="x vs. t")
sombrero.plot(t, y, label="x(t)", xlabel="t", ylabel="x", title="x vs. t")
```
